```python
from __future__ import print_function

import requests
from bs4 import BeautifulSoup
import os
```

```python
def file_links_filter(tag):
    """
    Tags filter. Return True for links that ends with 'pdf', 'htm' or 'txt'
    """
    if isinstance(tag, str):
        return tag.endswith('pdf')


def get_links(tags_list):
    return [WEB_ROOT + tag.attrs['href'] for tag in tags_list]


def download_file(file_link, folder):
    
    name = file_link.split('/')[-1]
    print('Connecting to ' + name)
    save_path = folder + name
        
    if (not os.path.isfile(SAVE_FOLDER + name)):
        
        try:
            file = requests.get(file_link).content
        except:
            print('Could not connect to ' +  name)
        
        print("Saving file:", save_path)
        
        try:
            with open(save_path, 'wb') as fp:
                fp.write(file)
                print('Saved file')
        except:
            print("An exception occurred")
    else:
        print("File found in directory" + name)
```

```python
WEB_ROOT = 'http://queries.gabs.co.za/'
SAVE_FOLDER = 'D:/001_Projects/08 - Bellville/Timetables/'  # directory in which files will be downloaded

r = requests.get("http://queries.gabs.co.za/timetables_pdf.asp?filter=AZ")

soup = BeautifulSoup(r.content, 'html.parser')

PDFs = soup.find_all('a', href=file_links_filter)  # css selector for all <a> inside <p id='archive'> tag
PDF_links = get_links(PDFs)

# make set to exclude duplicate links
PDF_links = set(PDF_links)

print("Got links")

for link in set(PDF_links):
    download_file(link, SAVE_FOLDER)
```
