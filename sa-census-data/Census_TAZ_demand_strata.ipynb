```python
import pandas as pd
```

```python
filename = 'Trip Generation.xls'
path ='D:/001_Projects/01 - GIPTN/04 - Macro Model/Census Export/%s' % (filename)
voertuigbewegings = pd.read_excel(path, nrows = 945, sheet_name = "Education", usecols = 10)
```

```python
final = pd.melt(voertuigbewegings, id_vars=['TAZ Number'], 
        var_name='Demand Strata', value_name='Number of Individuals')
final['Demand Strata'] = final['Demand Strata'].replace(['School.1', 'School.2'], 'School')
final['Demand Strata'] = final['Demand Strata'].replace(['Student.1', 'Student.2'], 'Student')
final['Demand Strata'] = final['Demand Strata'].replace(['Other.1', 'Other.2', 'Other.3'], 'Other')
result = final.groupby(by = ['TAZ Number', 'Demand Strata'])['Number of Individuals'].sum().reset_index()
result
```
