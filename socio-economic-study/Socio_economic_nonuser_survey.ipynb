```python
import pandas as pd
import glob
import numpy as np
import itertools
from ast import literal_eval
```

```python
def parse_data(path, sht):
    df = pd.read_excel(path, sheet_name = 2, header = 2, usecols ="O:BD")
    df = df.replace({'= Most important reason for not using the system':'', 
                     '= Less important reason for not using the system':'', 
                     '= Will not encourage me at all':'',
                     '= wWill not encourage me at all':'',
                     '= wWill most likely encourage me':'', 
                     ' most likely':'',  
                     ' less likely':'',
                     '= Will encourage me':'',
                     '= Heel belangrikste':'',
                     '= Minste belangrik':'',
                     '= heel belangrikste':'',
                     '= minste belangrik':'',
                     '= hoogs waarskynlik':'',
                     '= minste moontlik':''},
                    regex=True)
    
    #df = df.apply(literal_eval)
    
    return df
```

```python
path ='D:/001_Projects/01 - GIPTN/02 - Application development/05 - Socio-economic study/Non-User/'
rpath = 'D:/001_Projects/01 - GIPTN/02 - Application development/05 - Socio-economic study/Results/'

files = glob.glob(path +'[!~$]*')

df = []
for filename in files:
    df.append(parse_data(filename, 0))

df = pd.concat(df, axis=0, sort = False)

#df.to_csv(rpath + 'combined.csv')

pd.Series(df.columns.tolist()).to_csv(rpath + 'test.csv')

l = len(df.columns)

df = df.reset_index(drop=True)
```
