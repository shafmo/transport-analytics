```python
import pandas as pd
import numpy as np
```

```python
sheet_name = 'Heavy AM 2039'

mat = pd.read_excel('C:/Users/shafeeq.mollagee/OneDrive - Aurecon Group/TIA Microsimulations/01 - Cargo TIA/06 - Matrices/Demand Setup/Intermediate Matrices 2039.xlsx',
                   sheet_name = sheet_name)

mat = pd.melt(mat, id_vars=['Generation'], var_name='myVarname', value_name='myValname')
mat.columns = ['From', 'To', 'Count']

mat.To = mat.To.astype(float)
mat.To = mat.To.astype(int)

mat = mat.groupby(['From', 'To']).sum().reset_index()
mat = mat.pivot(index='From', columns='To', values='Count')
mat.values[[np.arange(mat.shape[0])]*2] = 0

mat.to_csv('C:/Users/shafeeq.mollagee/OneDrive - Aurecon Group/TIA Microsimulations/01 - Cargo TIA/06 - Matrices/Demand Setup/%s.csv' % (sheet_name))

mat.sum().sum()
```
