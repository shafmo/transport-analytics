```python
import pandas as pd
import numpy as np
```

```python
data_f = 'TAZ Development'
data_path = 'D:/001_Projects/01 - GIPTN/04 - Macro Model/Census Export/%s.xlsx' % (data_f)
```

```python
rates = pd.read_excel(data_path, usecols = 'A:J',sheet_name = 'Trip Generation Rates', nrows = 36)
taz = pd.read_excel(data_path, usecols = 'T', squeeze = True, sheet_name = 'Trip Generation Rates')
zoning = pd.read_excel(data_path, sheet_name = 'Intersected Pivot Values')

relation = pd.read_excel(data_path, usecols = 'L:R', sheet_name = 'Trip Generation Rates')

rates['AM1'] = rates[['AM Rate', 'Rate AM In']].product(axis=1)
rates['PM1'] = rates[['PM Rate', 'Rate PM In']].product(axis=1)
rates['AM'] = rates['AM1']/rates['Converted Unit']
rates['PM'] = rates['PM1']/rates['Converted Unit']

rates = rates[['Zoning', 'AM', 'PM']]

zoningm = pd.melt(zoning, id_vars=['TAZ'], var_name='zoning', value_name='area').dropna()
```

```python
relationf = pd.melt(relation, var_name='strat', value_name='zoning').dropna()
relationf = relationf[['zoning', 'strat']]
```

```python
df = zoningm.merge(rates, left_on='zoning', right_on='Zoning', how='outer').dropna()
df['AMA'] = df['AM']*df['area']
df['PMA'] = df['PM']*df['area']
df = df[['TAZ', 'zoning', 'AMA', 'PMA']]
```

```python
df1 = df.merge(relationf, left_on='zoning', right_on='zoning', how='outer').dropna()
df1.groupby(['TAZ', 'strat'])[['AMA', 'PMA']].sum()
table = pd.pivot_table(df1, values=['AMA', 'PMA'], index=['TAZ'], columns=['strat'], aggfunc=np.sum)
table
```

```python
data_r = 'Attraction'
data_path = 'D:/001_Projects/01 - GIPTN/04 - Macro Model/Census Export/%s.csv' % (data_r)
table.to_csv(data_path)
```
