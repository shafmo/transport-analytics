```python
import pandas as pd
import ntpath
import glob
```

```python
path ='D:/002_OtherWork/TOC Costing/Association Data/'
filenames = glob.glob(path + "/*.xls")

def read_sheet(sheet_name):
    dfs = []
    for filename in filenames:
        df = pd.read_excel(filename, sheet_name = sheet_name)
        df['Association'] = ntpath.basename(filename)
        dfs.append(df)

    return pd.concat(dfs, ignore_index=True)

sheet_name = 'Vehicles'

df = read_sheet(sheet_name).groupby(['Association', 'Year'])['Type'].count().reset_index()
df.to_excel("D:/002_OtherWork/TOC Costing/%s.xlsx" % (sheet_name), sheet_name=sheet_name)
df
```
